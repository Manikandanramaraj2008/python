```python
# inheritance function Aeroplane, cycle, Truck functions and object Boeing, Mack, TI created

class vehicle:
    def __init__(self,make,color):
        self.make = make
        self.color = color
        
    def moves(self): 
        print("Moves long...!!!")
        
        
    def get_model(self):
        print(f"i am made by {self.make} color of {self.color}.")
        
        
my_car = vehicle("TATA","BLUE")
my_car.get_model()


your_car = vehicle("Nissan","Gold")
print(your_car.make)

class Aeroplane(vehicle):  # inheritance function passing the initial class vehicle in to the new class Aeroplane
    def moves(self):       # using the existing function moves, but changing the printing content
        print("Flies along...!!!")  
        
class Truck(vehicle):  # can pass the initial class to n number of new classes can be created
    def moves(self):  # the properties can not be rewrite in code.
        print("giring along...!!!")
        
class cycle(vehicle):
            pass  #  no need def function smoves...!!pass code return to use the iniital print
        
   # create objects for truck, aeroplane and cycle.

Boeing = Aeroplane("Boeing", "yellow")
Mack = Truck("Mack","Pinnacle")
TI = cycle("TI","Herc")

Boeing.get_model()
Mack.moves()
TI.get_model()
TI.moves()
print(TI.make)
print(Mack.color)            
            
              
```

```python
# inheritance function additionally added ID to Aeroplane class

class vehicle:
    def __init__(self,make,color):
        self.make = make
        self.color = color
        
    def moves(self): 
        print("Moves long...!!!")
        
        
    def get_model(self):
        print(f"i am made by {self.make} color of {self.color}.")
        
        
my_car = vehicle("TATA","BLUE")
my_car.get_model()


your_car = vehicle("Nissan","Gold")
print(your_car.make)

class Aeroplane(vehicle):  # inheritance function passing the initial class vehicle in to the new class Aeroplane
# Adding initialisation of objects properties. use super function to avoid repitation of code to make and color.

    def __init__(self,make,color,fa_id): # initialisation
        super().__init__(make,color)  # use the vehicle function make and color 
        self.fa_id = fa_id  # id only added other properties kept as it is
                       
    def moves(self):       # using the existing function moves, but changing the printing content
        print("Flies along...!!!")  
        
class Truck(vehicle):  # can pass the initial class to n number of new classes can be created
    def moves(self):  # the properties can not be rewrite in code.
        print("giring along...!!!")
        
class cycle(vehicle):
            pass  #  no need def function smoves...!!pass code return to use the iniital print
        
   # create objects for truck, aeroplane and cycle.

Boeing = Aeroplane("Boeing", "yellow","fa253") # Id has added
Mack = Truck("Mack","Pinnacle")
TI = cycle("TI","Herc")

Boeing.get_model()
Boeing.moves()
print(Boeing.fa_id)
print(Boeing.color)
            
          
              
```

```python
# inheritance function all objects used in for loop getting results

class vehicle:
    def __init__(self,make,color):
        self.make = make
        self.color = color
        
    def moves(self): 
        print("Moves long...!!!")
        
        
    def get_model(self):
        print(f"i am made by {self.make} color of {self.color}.")
        
        
my_car = vehicle("TATA","BLUE")
my_car.get_model()


your_car = vehicle("Nissan","Gold")
print(your_car.make)

class Aeroplane(vehicle):  # inheritance function passing the initial class vehicle in to the new class Aeroplane
# Adding initialisation of objects properties. use super function to avoid repitation of code to make and color.

    def __init__(self,make,color,fa_id): # initialisation
        super().__init__(make,color)  # use the vehicle function make and color 
        self.fa_id = fa_id  # id only added other properties kept as it is
                       
    def moves(self):       # using the existing function moves, but changing the printing content
        print("Flies along...!!!")  
        
class Truck(vehicle):  # can pass the initial class to n number of new classes can be created
    def moves(self):  # the properties can not be rewrite in code.
        print("giring along...!!!")
        
class cycle(vehicle):
            pass  #  no need def function smoves...!!pass code return to use the iniital print
        
   # create objects for truck, aeroplane and cycle.

Boeing = Aeroplane("Boeing", "yellow","fa253") # Id has added
Mack = Truck("Mack","Pinnacle")
TI = cycle("TI","Herc")

for v in(my_car,your_car,Boeing,Mack,TI):
    v.get_model()
    v.moves()
    print(v.color)
    print(v.make)



            
          
              
```
